```python
# Lemmatization technique is steming. The output we will get after applying lemmatization technique is called Lemma, which is a
# root word rather than root stem, the output of steming technique. After lemmatization, we will get a valid word.
```

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
print(lemmatizer.lemmatize("cacti"))
print(lemmatizer.lemmatize("better",pos='a'))
```

```python
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```
